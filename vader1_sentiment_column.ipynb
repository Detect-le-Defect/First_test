```python
import pandas as pd
import numpy as np

import nltk
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
data = pd.read_csv('chatgreekgodx091421Clean.csv')
data.head()
```

```python
analyser = SentimentIntensityAnalyzer()

```

```python


new_emotes = {
    'omegalul': -2,
    'lilw': 0,
    'monkaw': -3.0,
    'waytoodank': 0,
    'ez': 2.0,
    'clap' : 3.0,
    'lul' : 0,
    'booba' : 1.0,
    'pepega' : -3.0,
    'feelsgoodman' : 3.0,
    'wutface' : -3.0,
    'sadge' : -4.0,
    'kreygasm' : 3.0,
    'kappa' : 0,
    'peeposad' : -4.0,
    'pogu' : 4.0,
    'truening' : 0,
    'monkas' : -2
}
```

```python
analyser.lexicon.update(new_emotes)
```

```python
data['scores'] = data['Cleanest_message'].apply(lambda clean_message:analyser.polarity_scores(clean_message))
```

```python


data['comp_score'] = data['scores'].apply(lambda score_dict: score_dict['compound'])
```

```python
data.head()
```

```python
data['sentiment']=''
data.loc[data.comp_score>0,'sentiment']='POSITIVE'
data.loc[data.comp_score==0,'sentiment']='NEUTRAL'
data.loc[data.comp_score<0,'sentiment']='NEGATIVE'
```
